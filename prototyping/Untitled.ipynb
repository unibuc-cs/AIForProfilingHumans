```python
from sklearn import mixture
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


df = pd.read_csv('weight-height.csv')
df.head
weights = df['Weight'].to_numpy()
print(weights)
plt.hist(weights, bins = 10, density=False)
min_weight = min(weights)
max_weight = max(weights)
values = list(range(int(min_weight), int(max_weight)))

mean = np.mean(weights)
std = np.std(weights)




distribution3 = mixture.GaussianMixture(n_components=7, covariance_type='full')
input_weights = np.array(weights).reshape(-1, 1)
noise = np.random.normal(mean/10, std/4, size=(input_weights.shape))
input_weights = np.concatenate((input_weights, input_weights/2 + noise), axis = 1)
print(input_weights.shape)

#print(input_weights)
distribution3 = distribution3.fit(input_weights)
#print(distribution2)

print(distribution3.predict_proba(input_weights))

print(distribution3.means_)
print(distribution3.covariances_)
```

```python
# A 2D case !

n_components = 1
distribution = mixture.GaussianMixture(n_components=5, covariance_type='full')
input_weights = np.array(weights).reshape(-1, 2)
#print(input_weights)
distribution = distribution.fit(input_weights)
#print(distribution2)

#print(distribution2.predict_proba(input_weights))
#print(distribution2.means_)
#print(distribution2.covariances_)

predictions = distribution.predict(input_weights)

print(input_weights.shape)
print(predictions.shape)
print(predictions.dtype)

numClasses = distribution.means_.shape[0]
#colors = cm.rainbow(numClasses)

from itertools import cycle
cycol = cycle('bgrcmk')
colormap  = np.array([next(cycol) for i in range(numClasses)])
print(colormap)

fig, ax = plt.subplots()

colorsToUse = colormap[predictions]
print(colorsToUse.shape, colorsToUse[:3])
print(input_weights.shape[0])
scatterRes = ax.scatter(input_weights[:,0], input_weights[:, 1], s = 1, c=colorsToUse, label=predictions,
            alpha=0.5)

# Draw the legend
classes = ['Type ' + str(i) for i in range(1, numClasses + 1)]
class_colours = colormap
recs = []
for i in range(0,len(class_colours)):
    recs.append(mpatches.Rectangle((0,0),1,1,fc=class_colours[i]))
plt.legend(recs,classes,loc=4)

#ax.legend()

#legend1 = ax.legend(*scatterRes.legend_elements(),
#                    loc="lower left", title="Classes")
#ax.add_artist(legend1)

#print(distribution.means_)
#ax.scatter(distribution.means_[:, 0], distribution.means_[:, 1], s=200, c=colormap)


```
